**Bivariate Anaysis**. 

Now, let's make a plot to see degree of correlated column.

```python
water_g = sns.PairGrid(water_df, hue ="Potability")
water_g.map_upper(plt.scatter,alpha = 0.5)
water_g.map_lower(sns.kdeplot)
water_g.add_legend()
```

Above scatter and kde plot shows every combination for pair of columns. We can see that level of correlation is very low in every plot.

Let's quantify the correlation between columns using heatmap graph.

```python
sns.heatmap(water_df.corr(), annot=True, fmt='.2f', linewidths=2, cmap="Blues")
```

From above heatmap, we can see that Potability column has a miniature positive correlation with Turbidity,Trihalomethanes, Chloramines, Solids, and ph.  
**Hypothesis**   
1. After looking into the values from heatmap, we can assume that each columns individually have very less information to classify the water as a 0 or 1 label.

Now, lets look into the box plots.

```python
# Boxplot of Potability and ph
sns.boxplot(x='Potability', y='ph', data = water_df)
```

From above boxplot, we can see that median, Q1 and Q3 are overlapping for both Potable labels with respect to ph values. Hence it is hard to distinguish Potability solely based on ph value.

```python
# Boxplot of Potability and Hardness
sns.boxplot(x='Potability', y='Hardness', data = water_df)
```

In this boxplot also, we have almost overlapping statistics for potability with respect to Hardness.  
Hence, our hypotheis of each columns carrying very minute information seems to be correct based on the box plots analysis.

**Conclusion/Observations**

1. From univariate analysis, we can conclude that there is not a line that can easily separate the labels using individual columns only.  
2. From bivariate analysis, we can conclude that there is almost zero correlation among columns. In this case, predicting labels using single column only is hard as columns have very small correlation with the target label.
